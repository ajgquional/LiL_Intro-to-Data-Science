# Question 1: What were the 5 most popular baby names in 2014 in the US?

```python
us_babies
```

```python
us_babies['Year']
```

```python
us_babies['Year'] == 2014
```

```python
us_babies_2014 = us_babies.loc[us_babies['Year'] == 2014, :]
```

```python
us_babies_2014
```

```python
sorted_us_2014 = us_babies_2014.sort_values('Count', ascending = False)
```

```python
sorted_us_2014
```

```python
sorted_us_2014.iloc[0:5]
```
