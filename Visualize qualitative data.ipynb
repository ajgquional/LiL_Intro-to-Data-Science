```python
import pandas as pd
import matplotlib.pyplot as plt # to create figures and plots
import seaborn as sn # to create attractive and interactive visualizations
```

```python
listings = pd.read_csv('Airbnb_NYC_2019.csv')
```

```python
listings
```

```python
# creating bar chart containing Airbnb listings in each neighborhood group of NYC

sn.countplot(x = 'neighbourhood_group', data = listings) # bar chart
plt.show()
```

```python
# creating bar chart displaying the average price of Airbnb listings in each neighborhood group in NYC in the listings data frame

sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings) # bar chart
plt.show()

# black lines in each bar show confidence intervals generated by the barplot method
```

```python
# within the confidence intervals

sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings, ci = False)
plt.show()
```
